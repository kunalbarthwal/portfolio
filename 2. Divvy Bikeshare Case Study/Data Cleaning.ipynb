### 3) Cleaning the data
We found several areas where we can improve the integrity of our data for ease of use:
-  Datatype of _started_at_ and _ended_at_ columns can be changed to _datetime_ format
-  Checking for unique values
-  Removing duplicates
-  Creating new columns
-  Renaming columns 

```python
new_df['started_at'] = pd.to_datetime(new_df['started_at'],errors='coerce')
```

```python
new_df['ended_at'] = pd.to_datetime(new_df['ended_at'],errors='coerce')
```

```python
new_df.info()
```

```python
new_df['ride_id'].value_counts()
```

```python
new_df = new_df.drop_duplicates(subset=['ride_id'])
```

```python
new_df['rideable_type'].unique()
```

```python
new_df["start_station"] = new_df['start_station_name'] +" - "+ new_df["start_station_id"]
```

```python
new_df["end_station"] = new_df['end_station_name'] +" - "+ new_df["end_station_id"]
```

```python
new_df['member_casual'].unique()
```

```python
new_df['member_casual'].value_counts()
```

```python
new_df = new_df.rename(columns={'rideable_type': 'bike_type', 'started_at': 'start_time', 'ended_at': 'end_time', 'member_casual': 'membership_type',})
```

```python
new_df['day'] = new_df["day"].map(str)
```

```python
new_df["day"] = new_df["day"].replace(['1', '2', '3', '4', '5', '6', '7'], ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'])
```

```python
new_df
```
